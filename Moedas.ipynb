```python
from bs4 import BeautifulSoup
import requests
```

```python
import pandas as pd
from datetime import datetime
```

```python
res = []
```

```python
url = 'https://www.infomoney.com.br/ferramentas/cambio/'
response = requests.get(url)
html = response.content

soup = BeautifulSoup(html, 'html.parser')
table = soup.find('table', attrs={'class':'default-table'})
table_rows = table.find_all('tr')

#=================================================================================#

for tr in table_rows:
    date = datetime.now()
    now = date.strftime("%d/%m/%Y %H:%M:%S")
    td = tr.find_all('td')
    row = [tr.text for tr in td]
    row = row + [now]
    if row:
        res.append(row)
        
df = pd.DataFrame(res, columns=['Moeda', '', 'Compra', 'Venda', 'VAR(%)', 'Data'])
df = df.dropna()

#=================================================================================#

df
```
